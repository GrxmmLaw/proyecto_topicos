```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the datasets
df_distintos = pd.read_csv('resultado_distintos_SCALED - Copy.csv')
df_especies = pd.read_csv('resultado_especies_SCALED.csv')
df_cepas = pd.read_csv('resultado_cepas_SCALED.csv')

# Function to calculate statistics
def calculate_stats(df, name):
    # Split into calibration (first 5 windows) and monitoring
    calibration = df[df['estado'] == 'CALIBRANDO']
    monitoring = df[df['estado'] == 'MONITOREO']
    
    baseline_noise = calibration['rareza_rho'].mean()
    noise_std = calibration['rareza_rho'].std()
    threshold_final = df['umbral_tau'].iloc[-1]
    
    # Peak detection in monitoring phase
    peak_rho = monitoring['rareza_rho'].max()
    alerts_count = monitoring['alerta'].sum()
    
    # Success check
    detection_success = alerts_count > 0
    
    return {
        "Escenario": name,
        "Ruido Base (Mean Rho)": baseline_noise,
        "Inestabilidad (Std Rho)": noise_std,
        "Umbral (Tau)": threshold_final,
        "Pico Máximo (Max Rho)": peak_rho,
        "Alertas Detectadas": alerts_count,
        "¿Detectó Anomalía?": "SÍ" if detection_success else "NO"
    }

# Calculate stats for all
stats_list = [
    calculate_stats(df_distintos, "Distintos (Diverso)"),
    calculate_stats(df_especies, "Especies (Medio)"),
    calculate_stats(df_cepas, "Cepas (Similar)")
]

stats_df = pd.DataFrame(stats_list)

# Plotting
fig, axes = plt.subplots(3, 1, figsize=(12, 12), sharex=True)
fig.suptitle('Resultados Experimentales: Escala Masiva (5M reads, Ventana 200k)', fontsize=16)

# Helper for plotting
def plot_scenario(ax, df, title, color):
    ax.plot(df['ventana'], df['rareza_rho'], label='Rareza (Rho)', color=color, linewidth=2)
    ax.plot(df['ventana'], df['umbral_tau'], label='Umbral Adaptativo (Tau)', color='black', linestyle='--', linewidth=1.5)
    

    
    ax.set_title(title)
    ax.set_ylabel('Índice de Rareza (Rho)')
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)

plot_scenario(axes[0], df_distintos, "Escenario 1: Genomas Distintos (Catálogo Masivo)", 'tab:blue')
plot_scenario(axes[1], df_especies, "Escenario 2: Especies Diferentes (Catálogo Medio)", 'tab:orange')
plot_scenario(axes[2], df_cepas, "Escenario 3: Cepas Similares (Catálogo Pequeño)", 'tab:green')

axes[2].set_xlabel('Número de Ventana')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('analisis_resultados_scaled.png')

print("Estadísticas del Experimento:")
print(stats_df)
```
